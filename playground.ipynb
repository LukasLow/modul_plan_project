```python
import tabula

df = tabula.read_pdf("ModuleDS7.pdf", multiple_tables=True, pages= "all")

with open("output2.txt", "w") as f:
    for i in range(len(df)):
        f.write("\ntable begin\n")
        f.write(df[i].to_string())
        f.write("\ntable end\n")
```
